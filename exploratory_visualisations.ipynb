```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_apartment_data.csv')

df.head()
```

```python
df.describe()
```

```python
df_outliers_removed = pd.read_csv('cleaned_apartment_data_outliers_removed.csv')

df_outliers_removed.head()
```

```python
df_outliers_removed['price'].describe()
```

```python
df['price'].quantile(0.98)
```

```python
df['price'].groupby(df['state']).agg('mean')
```

```python
sns.relplot(x=df_outliers_removed['bedrooms'], y=df_outliers_removed['price'], kind='scatter')
plt.ylim(0, 5000)
plt.show()
```

```python
sns.histplot(data=df_outliers_removed, x='price')
plt.xlim(0, 5000)
plt.show()
```

```python
vars_list = ['bathrooms', 'bedrooms', 'price', 'square_feet']

sns.pairplot(df_outliers_removed, vars=vars_list)
```

```python
sns.displot(
    data=df_outliers_removed, 
    x='price', 
    col='state', 
    col_wrap=5, 
    kind='hist', 
    kde=True,
    binrange=(0,5000))
plt.xlim(0, 5000)
plt.show()
```
